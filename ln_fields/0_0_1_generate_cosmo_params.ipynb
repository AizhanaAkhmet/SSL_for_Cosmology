```python
import os, time
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import qmc
```

## Sample data

```python
seed = 1 # random seed 
dim  = 2 # number of dims
num_samples_k = 10 # of data points, x1000
num_samples = num_samples_k*1000 

# sample with LH
sampler = qmc.LatinHypercube(d=dim, seed=seed) 
sample = sampler.random(n=num_samples)
print('Discrepancy: ', qmc.discrepancy(sample))

# scale parameters
l_bounds = [0.15, 0.65]
u_bounds = [0.45, 0.95]
cosmo_params = qmc.scale(sample, l_bounds, u_bounds)

# plot sampled points
plt.figure()
plt.xlabel(r'$\Omega_M$')
plt.ylabel(r'$\sigma_8$')
plt.axhline(0.8, c = 'k')
plt.axvline(0.3, c = 'k')
plt.scatter(cosmo_params[:, 0], cosmo_params[:, 1])
plt.show()
```
